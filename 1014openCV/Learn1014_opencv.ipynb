```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=np.full((300, 500, 3), (255,255,255), np.uint8) #先做一張白色底圖,300*500,色彩是彩色
cv2.line(m1, (20,20), (20,200), (0,0,255), 2) #畫一條線:變數,起點,終點,線條顏色,線條粗細

cv2.rectangle(m1, (40,40), (150,150), (0,255,0), 2) #畫一個方形:變數,起點,終點,線條顏色,線條粗細(-1就是實心)
cv2.rectangle(m1, (200,40), (300,150), (0,255,0), -1)

cv2.circle(m1, (100,230), 30, (255,0,0), 2) #畫一個圓形:變數,起點,半徑,線條顏色,線條粗細(-1就是實心)
cv2.circle(m1, (200,230), 30, (255,0,0), -1)

cv2_imshow(m1)
```

```python
from PIL import ImageFont, ImageDraw, Image #需使用PIL函式庫的Image,ImageFont,ImageDraw模組, >>>可以載入字型
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

#在colab裡面,如果你沒有在這段裡面宣告m1,會自動繼承前面幾段用到的m1
# m1=np.full((300, 500, 3), (255,255,255), np.uint8) #用底圖來畫
m1=cv2.imread("1.jpg", 1) #用背景圖來畫

p1=Image.fromarray(m1) #這裡用來將m1(cv2函式庫)轉成p1(PIL函式庫)

ImageDraw.Draw(p1).text((30,30), "輸入你要加的文字,aabb123 * / ★", (0,0,255), ImageFont.truetype("kaiu.ttf",40))  #位置會從30,30,文字貼齊左上開始
#加入字體,(已轉呈PIL底圖的參數).text(文字起始位置) , "你要加的字" , 字體顏色 , ("字體名稱",字體大小)

m2=np.array(p1) #將p1轉回cv2函式庫,存到m2裡 , 因為PIL是一個封閉函式庫,要show還是要透過cv2的函示庫

cv2_imshow(m2)
```

```python
from PIL import ImageFont, ImageDraw, Image
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time #因為要使用到time.sleep , 要import time函式庫

#圓形動畫
m1=np.full((500, 500, 3), (255,255,255), np.uint8) #做一個500*500的白色底
for w in range(5,250,30): #讓w做為半徑變數,從5開始,每跑一次迴圈就+15,直到250
  dp.clear_output(True) #清除畫面(等待新輸出) #是用來把原畫面清掉的,這樣就會在同一張圖片上做變化 #用Ture比較順,Flase會像圖片切換一樣 ###本地端不用
  cv2.circle(m1, (w,w), w+20 , (255,w,0),5) #畫一個圓,圓心位置從(250,250)開始,w為半徑,顏色(255,0,0),-1表示實心 #這裡的變數可以玩
  cv2_imshow(m1) #每次迴圈跑一個新圓,就show一次   #本地端用cv2_imshow("win1",m1)
  time.sleep(0.5) #讓圓停個0.3秒,才不會一下子跑完,比較有動畫的流暢感 ###本地端用cv2.waitKey(300)
  ###網路效能有限,所以速度上不會像本機上這麼順暢
```
