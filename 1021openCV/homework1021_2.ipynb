```python
from PIL import ImageFont, ImageDraw, Image
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time

m1=cv2.imread("h2.png",1)

# cv2_imshow(m1)
m2=np.full((m1.shape[0],m1.shape[1], 3), (0,0,255), np.uint8) #先拔掉變成黑
# cv2_imshow(m2)
m3=cv2.absdiff(m1,m2) #為何不能用一般減法?>>>因為全部的紅色都變成0了,這樣怎麼乘,矩陣內所有數值的最後一格都會變0
cv2_imshow(m3)
m4=np.full((m1.shape[0],m1.shape[1], 3), (255,255,255), np.uint8)
m5=cv2.multiply(m3,m4)
# cv2_imshow(m5)
m6=m5[:,:,2]
cv2_imshow(m6)


# #最後試法(有超前進度指令):直接用inRange把紅色挑出來,在用not運算翻轉,就出來了
# m6=cv2.inRange(m1,(0,0,255), (0,0,255))
# # cv2_imshow(m6)
# m7=cv2.bitwise_not(m6)
# cv2_imshow(m7)
# cv2.imwrite("1021hw_ans.jpg", m7, [cv2.IMWRITE_JPEG_QUALITY,100])
```
