```python
#寫一個可以動畫的程式。
#程式中有一個方形會在左右兩端來回的跑
#直到按下任意按鍵為止，否則程式不會結束
# !pip install getch #>>>限windows,colab是Linux
# import msvcrt
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time

x = 0
p1=cv2.VideoWriter("1.mp4", cv2.VideoWriter_fourcc(*'MJPG'), 24, (500,300)) #這裡是寬,長

m1=np.full((300, 500, 3), (255,255,255), np.uint8) #先做一張白色底圖,300*500,色彩是彩色
i = 0
while i < 3:
  if x == 0:
    print(f'以上是第{i+1}次')
    i = i+1
    for x in range(0,m1.shape[1]-100+1,50):
      y = x+100
      dp.clear_output(True)
      m1=np.full((300, 500, 3), (255,255,255), np.uint8) #先做一張白色底圖,300*500,色彩是彩色
      cv2.rectangle(m1, (x,100), (y,200), (0,0,0), -1) #畫一個方形:變數,起點,終點,線條顏色,線條粗細(-1就是實心)
      p1.write(m1) #寫入影片影格 ##因為他是寫入影格,不是開始錄影,所以要寫在for迴圈裡面,記錄每一次變化
      time.sleep(0.3)
      cv2_imshow(m1)
      print(f"for++{x},{y}")
  if x >= 400:
    for x in range(m1.shape[1]-100,0-1,-50):
      y = x+100
      dp.clear_output(True)
      m1=np.full((300, 500, 3), (255,255,255), np.uint8) #先做一張白色底圖,300*500,色彩是彩色
      cv2.rectangle(m1, (x,100), (y,200), (0,0,0), -1) #畫一個方形:變數,起點,終點,線條顏色,線條粗細(-1就是實心)
      p1.write(m1) #寫入影片影格 ##因為他是寫入影格,不是開始錄影,所以要寫在for迴圈裡面,記錄每一次變化
      time.sleep(0.3)
      cv2_imshow(m1)
      print(f"for--{x},{y}")
p1.release() #釋放控制變數-要放,不然可能會壞掉,代表沒有結束
```
